-sandbox ## Data Preprocessing

```
from pyspark.sql.functions import col

moviesDF = moviesDF.select(feats_i).dropna()
exprs = [col(column).alias(column.replace(' ', '_')) for column in feats_i]
display(moviesDF)
```

```
# String Indexers
from pyspark.ml.feature import StringIndexer

indexers = []
cat_cols = []
vec_cols = []
for i, col_name in enumerate(moviesDF.columns):
  #if moviesDF.dtypes[i][1] == 'string':
  if True:
    cat_col = 'cat_' + col_name
    vec_col = 'vec_' + col_name
    cat_cols.append(cat_col)
    vec_cols.append(vec_col)
    indexers.append(StringIndexer(inputCol=col_name, outputCol=cat_col, handleInvalid='skip'))
```

```
# One-Hot Encoder
from pyspark.ml.feature import OneHotEncoderEstimator

encoder = OneHotEncoderEstimator(
  inputCols=cat_cols,
  outputCols=vec_cols
)
```

```
# Vector Assembler

from pyspark.ml.feature import VectorAssembler

featureCols = moviesDF.columns
assembler = VectorAssembler(inputCols=featureCols, outputCol='features')
```

```
# Pipeline
from pyspark.ml import Pipeline

pipeline = Pipeline(stages=
  indexers + [ 
  encoder,
  assembler
])

pipelineModel = pipeline.fit(moviesDF)
transformedDF = pipelineModel.transform(moviesDF)

display(transformedDF)
```

```
import numpy as np
from numpy.random import seed
from numpy.random import randn
from numpy import percentile
import pandas as pd

# identify and drop outliers with interquartile range
def drop_outliers(data):
  
  # calculate interquartile range
  q25, q75 = percentile(data, 25), percentile(data, 75)
  iqr = q75 - q25

  # calculate the outlier cutoff
  cut_off = iqr * 1.5
  lower, upper = q25 - cut_off, q75 + cut_off
  
  outliers = [x for x in data if (not np.isnan(x)) and (x < lower or upper < x)]
  outliers_removed = data
  count = 0
  for value in data:
    if np.isnan(value) or (lower < value < upper):
      outliers_removed[count] = value
    else:
      outliers_removed[count] = np.nan
    count += 1

  return outliers_removed

# create new dataframe with no outliers
df = pd.DataFrame(columns=moviesDF.columns)
for i, col_name in enumerate(moviesDF.columns):
  if moviesDF.dtypes[i][1] == 'int' or moviesDF.dtypes[i][1] == 'double' or moviesDF.dtypes[i][1] == 'long':
    df[col_name] = pd.Series(drop_outliers(moviesDF.select(col_name).collect()))

df.dropna(inplace=True)
print(df)
```

-sandbox ## Model Training and Testing

```
seed = 1
(testDF, trainDF) = moviesDF.randomSplit((0.2, 0.8), seed=seed)

print(testDF.count(), trainDF.count())
```

```
feat_i = feats_i[0] # an example
```

```
from pyspark.ml.linalg import Vectors, VectorUDT

spark.udf.register("oneElementVec", lambda d: Vectors.dense([d]), returnType=VectorUDT())
expr = 'oneElementVec(' + feat_i + ') as feat'
trainDF = trainDF.selectExpr('imdb_score', expr)
testDF = testDF.selectExpr('imdb_score', expr)
predictionsDFs = []
```

```
from pyspark.ml.regression import LinearRegression

# Model
lr = LinearRegression(labelCol='imdb_score', featuresCol='feat')
piplineFull = Pipeline(stages=[lr])
lrModel = piplineFull.fit(trainDF)

# Predictions
predictionsLrDF = lrModel.transform(testDF)
predictionsDFs.append(predictionsLrDF)

# To see how it performed on the non-zero labels, sort it in decreasing order
display(predictionsLrDF.orderBy(predictionsLrDF.imdb_score.desc()))
```

```
from pyspark.ml.regression import RandomForestRegressor

# Model
rf = RandomForestRegressor(labelCol='imdb_score', featuresCol='feat')
piplineFull = Pipeline(stages=[rf])
rfModel = piplineFull.fit(trainDF)

# Predictions
predictionsRfDF = rfModel.transform(testDF)
predictionsDFs.append(predictionsRfDF)

# To see how it performed on the non-zero labels, sort it in decreasing order
display(predictionsRfDF.orderBy(predictionsRfDF.imdb_score.desc()))
```

```
from pyspark.ml.regression import GBTRegressor

# Model
gb = GBTRegressor(labelCol='imdb_score', featuresCol='feat')
piplineFull = Pipeline(stages=[gb])
gbModel = piplineFull.fit(trainDF)

# Predictions
predictionsGbDF = gbModel.transform(testDF)
predictionsDFs.append(predictionsGbDF)

# To see how it performed on the non-zero labels, sort it in decreasing order
display(predictionsGbDF.orderBy(predictionsGbDF.imdb_score.desc()))
```

```
from pyspark.ml.evaluation import RegressionEvaluator

metricNames = ['rmse', 'r2', 'mse', 'mae']
evaluator = RegressionEvaluator(labelCol='imdb_score')
for i, el in enumerate(predictionsDFs):
  print('\nAlgorithm #' + repr(i + 1))
  for metricName in metricNames:
    print(metricName + ' = ' + str(evaluator.setMetricName(metricName).evaluate(el)))
```
